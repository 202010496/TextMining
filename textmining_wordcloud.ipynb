```python
# 텍스트 분석 순서
# 문장 → 단어 → 키워드 → 필터링 → 집계 → 시각화
# 문장을 공백 단위로 분할하여 단어로 구성
# 사전과 대입하여 키워드 추출
# 팔요없는 단어 필터링
# 결과 집계
# 워드클라우드의 시각화 

# Step1. 필요한 모듈을 실행함
# 기본 준비 사항
from konlpy.tag import Kkma
import matplotlib.pyplot as plt  # pip 먼저 해야함
from matplotlib import font_manager, rc  # pip 먼저 해야함
from wordcloud import WordCloud   # pip 먼저 해야 함

# wordcloud 설치할 때, setuptools 에러나면, 아래 코드 실행 후 재실행
# pip install -U pip setuptools

from collections import Counter
import numpy as np
Kkma = Kkma()
```

```python
# Step2. 파일을 불러와서 형태소 분석하기
data1 = open("C:\\Users\\82109\\Documents\\카카오톡 받은 파일\\word.txt", 'rt', encoding='UTF8').read()
data1
data2 = Kkma.nouns(data1)
data3 = Counter(data2)
```

```python
# Step3. 불용어 제거하기
stop_words = open("C:\\Users\\82109\\Desktop\\파이썬을 활용한 데이터분석 및 시각화_실습(전체)v2\\7차시\\stop_words.txt").read()
data3 = [ each_word for each_word in data2
            if each_word not in stop_words ]

data3

# 1글자 이하이거나 10글자 이상인 단어 삭제하기
data4=[ ]
for i in range(0, len(data3)) :
    if len(data3[i]) >= 2 | len(data3[i]) <= 10 :
        data4.append(data3[i])
        
# 삭제하고 싶은 단어를 메모장에 정리 → 파일 안에 단어 일괄 삭제
# 불러온 단어가 모두 없어질 때까지 for문 반복

# 불용어 필터링 방법 : 1. 메모장에 불용어 제거 / 2. 글자 수 지정

# 문장 분석의 원리: 띄어쓰기 기준으로 단어로 만들어 분석

# 글자수의 제약이 필요
```

```python
# Step4. 단어 별 빈도수 집계하기
data5 = Counter(data4)
data6 = data5.most_common(200)
tmp_data = dict(data6)
```

```python
# Step5. 워드 클라우드 그리기
# 언급 빈도수에 따라 글자 크기를 설정
# 한글 폰트를 기본적으로 지원하지 않음
wordcloud = WordCloud(font_path = 'C:\\Users\\82109\\NanumFontSetup_TTF_BARUNPEN\\NanumBarunpenB.ttf',
                      relative_scaling = 1, 
                     background_color='white',
                     ).generate_from_frequencies(tmp_data)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.axis('on')
plt.show()
```
